# Julia implementation of rolling dice

[Another Math.SE question about rolling dice](https://math.stackexchange.com/q/2644796/290189).  This time, $\mathbb{E}[X_{\max}]$ is asked.  OP showed calculations for $\mathbb{E}[X_{\min}]$.

```julia
sample = [(i,j,k,l) for i in 1:6, j in 1:6, k in 1:6, l in 1:6]
```

```julia
Xmin = [minimum(roll) for roll in sample]
```

```julia
sum(Xmin) / 6^4  # 2275 // 1296
```

Let's return to the question.

```julia
Xmax = [maximum(roll) for roll in sample]
```

```julia
sum(Xmax)
```

```julia
sum(Xmax) // 6^4, sum(Xmax) / 6^4
```
